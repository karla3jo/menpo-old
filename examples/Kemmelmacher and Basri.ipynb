# Problem Definition
We wish to recover the shape from a single image given a reference shape $(z_{ref})$ and albedo $(\rho_{ref})$. The image irradiance function is defined as:

$$
I(x,y) = \rho (x,y) R(x,y)
$$

where the reflectance function $R(x,y)$ can be defined in terms of spherical harmonics as:

$$
R(n(x,y); \rho (x,y), l) = l^T Y(n(x,y))
$$

where $n(x,y)$ is the normal at pixel $(x,y)$, $\rho (x,y)$ the albedo and $l$ the 4-component vector of harmonic expansion coefficients. 

$Y$ is represents the **first order** spherical harmonic function for each normal, defined in vector form as:

$$
Y(n) = (1, n_x, n_y, n_z)^T
$$

In order to preserve discontinuties in the reference model, we define our optimisation problem as in terms of the shape and albedo differences:

$$
d_z (x,y) = z(x,y) - z_{ref}(x,y) \\
d_p (x,y) = p(x,y) - p_{ref}(x,y)
$$

and the optimisation function as:

$$
\min_{l,\rho,z} \sum_{(x,y) \in \Omega} \left( I(x,y) - \rho (x,y) l^T Y(n(x,y)) \right)^2 + \lambda_1 (\nabla G \ast d_z)^2 + \lambda_2 (\nabla G \ast d_p)^2
$$

We look to solve this optimisation function by seperating the three unknowns, $l$, $\rho$ and $z$.

# Creating a reference albedo map from a known light direction
We assume Lambertian reflectance, and thus that the image has been lit by a single dominant light source placed at infinity. Thus, the intensity of the image is defined as:

$$
I(x,y) = \rho (x,y) \max(n(x,y)^T s, 0)
$$

where $I$ is the pixel intensity, and $s$ is the direction vector of the light. Therefore, given an intensity image, a reference shape and light direction, we can uniquely determine $\rho (x,y)$ as:

$$
\rho (x,y) = \frac{I(x,y)}{n(x,y)^T s}
$$

# Step 1: Solving for unknown lighting
Given that we have a reference shape (and thus normal) and albedo, we can solve for the lighting parameters as follows:
$$
\min_{l} \sum_{(x,y) \in \Omega} \left( I(x,y) - \rho_{ref}(x,y) l^T \hat{n}_{ref}(x,y) \right)^2
$$

where $\hat{n} = (1, n_x, n_y, n_z)^T$.

Since we only have four unknowns, this represents a highly overdetermined linear least squares problem. We begin by vectorizing the function:

$$
\min \| I - P_{ref} \circ \hat{N_{ref}}^T l \|^2
$$

where $I$ has been concatenated in to a single column vector $(x_1, y_1, \ldots, x_N, y_N)$, $P_{ref}$ represents the concatenated column vector of $\rho_{ref}$ values and $\hat{N_{ref}}$ is a matrix of the form:

$$
\begin{pmatrix}
    1 & n^x_1 & n^y_1 & n^z_1 \\
    \vdots & \vdots & \vdots & \vdots \\
    1 & n^x_N & n^y_N & n^z_N
\end{pmatrix}
$$

Now, this optimisation function is minimised when $\frac{\partial f}{\partial l} = 0$. Therefore, by taking the derivative, we arrive at the solution:

$$
I - (P \circ \hat{N} l)(p \circ \hat{N}^T) = 0 \\
P \circ \hat{N}^T I - P \circ \hat{N}^T P \circ \hat{N} l = 0 \\
l = \left[(P \circ \hat{N})^T (P \circ \hat{N}) \right]^{-1} (P \circ \hat{N})^T I
$$

Given that our system is overdetermined, we note that $\left[(P \circ \hat{N})^T (P \circ \hat{N}) \right]^{-1} (P \circ \hat{N})^T$ is equivalent to the pseudo-inverse of $P \circ \hat{N}$.

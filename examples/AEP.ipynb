```python
def cart2sph(x, y, z):
    xy = np.power(x, 2) + np.power(y, 2)
    # for elevation angle defined from XY-plane up
    return np.vstack([np.arctan2(y, x), np.arctan2(z, np.sqrt(xy))]).T

def sph2cart(azimuth, elevation, r):
    cart = np.empty([azimuth.shape[0], 3])
    azi_cos = np.cos(azimuth)
    ele_cos = np.cos(elevation)
    azi_sin = np.sin(azimuth)
    ele_sin = np.sin(elevation)
    cart[:, 0] = r * ele_cos * azi_cos
    cart[:, 1] = r * ele_cos * azi_sin
    cart[:, 2] = r * ele_sin
    return cart
```

```python
def expmap_smith(U, bp):
    long0lat1 = cart2sph(-bp[:, 1], bp[:, 0], bp[:, 2])
    long0lat1[:, 0] += np.pi
 
    rho = np.sqrt(U[:, 0] ** 2 + U[:, 1] ** 2)
    Z = np.exp(np.arctan2(U[:, 1], U[:, 0]) * 1j)
    V1 = rho * np.real(Z)
    V2 = rho * np.imag(Z)
    # To prevent divide by 0 warnings when rho is 0
    ir = rho == 0
    eps = np.spacing(1)
    rho[ir] = eps
    c = rho
    
    c[ir] = eps
    
    Y = np.arcsin(np.cos(c) * np.sin(long0lat1[:, 1]) + np.cos(long0lat1[:, 1]) * np.sin(c) * V2 / rho)
    X = long0lat1[:, 0] + np.arctan2(V1 * np.sin(c), np.cos(long0lat1[:, 1]) * np.cos(c) * rho - np.sin(long0lat1[:, 1]) * V2 * np.sin(c))
    
    longlat = np.empty([X.shape[0], 2])
    longlat[:, 0] = np.real(X)
    longlat[:, 1] = np.real(Y)
    
    N = sph2cart(longlat[:, 0] - np.pi, longlat[:, 1], np.ones_like(longlat[:, 1]))
    # Swap x and y axes
    N[:, [0, 1]] = N[:, [1, 0]]
    N[:, 1] = -N[:, 1]
    
    return N
 
def logmap_smith(N, bp):
    longlat = cart2sph(-N[:, 1], N[:, 0], N[:, 2])
    longlat[:, 0] += np.pi
    
    long0lat1 = cart2sph(-bp[:, 1], bp[:, 0], bp[:, 2]) 
    long0lat1[:, 0] += np.pi
    
    c = np.arccos(np.sin(long0lat1[:, 1]) * np.sin(longlat[:, 1]) + np.cos(long0lat1[:, 1]) * np.cos(longlat[:, 1]) * np.cos(longlat[:, 0] - long0lat1[:, 0]))
    
    k = 1.0 / np.sin(c)
    k = np.nan_to_num(k)
    k = c * k
    
    U = np.empty([N.shape[0], 2])
    U[:, 0] = k * (np.cos(longlat[:, 1]) * np.sin(longlat[:, 0] - long0lat1[:, 0]))
    U[:, 1] = k * (np.cos(long0lat1[:, 1]) * np.sin(longlat[:, 1]) - np.sin(long0lat1[:, 1]) * np.cos(longlat[:, 1]) * np.cos(longlat[:, 0] - long0lat1[:, 0]))
                                                                                                           
    return U
```

```python
def logmap_snape(normals, mean_normal):
    if normals.shape[1] != 3:
        raise DimensionalityError('Normal vector is expected to contain normals')
    # If we've been passed a single vector to map, then add the extra axis
    if len(normals) == 3:
        vector_count = normals.shape[2]
    else:
        normals = normals[..., None]
        vector_count = 1
        
    # from x,y,z to x,y
    N = normals.shape[0]
    out = np.zeros([N, 2, vector_count])
    
    # Ensure unit normals
    vec_mean_normals = normalise_vector(mean_normal)
    
    thetaav = elevation(vec_mean_normals[:, 2])
    phiav = azimuth(vec_mean_normals[:, 0], vec_mean_normals[:, 1])
    phiav, thetaav = clip_angles(phiav, thetaav)
    
    for i in xrange(vector_count):
        # Get current vector
        kset = normals[:, :, i]
        projected = np.zeros([N, 2])
    
        # find any zero normals as they present a real problem
        # the column indicies are the same in both sets of data
        zero_indices = np.sum(np.abs(kset), axis=1) == 0.0
        
        thetak = elevation(kset[:, 2])
        phik = azimuth(kset[:, 0], kset[:, 1])
        phik, thetak = clip_angles(phik, thetak)
        
        # cos(c) = sin(thetaav) * sin(thetak) + cos(thetaav) * cos(thetak) * cos[phik - phiav]
        cosc = np.sin(thetaav) * np.sin(thetak) + np.cos(thetaav) * np.cos(thetak) * np.cos(phik - phiav)
        c = np.arccos(cosc);
        # kprime = c / sin(c)
        kprime = c / np.sin(c)
        
        # xs = kprime * cos(thetak) * sin[phik - phiav]
        projected[:, 0] = kprime * np.cos(thetak) * np.sin(phik - phiav)
        # ys = kprime * (cos(thetaav) * sin(phik) - sin(thetaav) * cos(thetak) * cos[phik - phiav]
        projected[:, 1] = kprime * (np.cos(thetaav) * np.sin(thetak) - np.sin(thetaav) * np.cos(thetak) * np.cos(phik - phiav))
    
        # reset the zero normals back to 0
        projected[:, zero_indices] = 0.0
        
        # reshape back to column vector
        out[:, :, i] = projected
        
    return out
        
def expmap_snape(projections, mean_normal):
    if projections.shape[1] != 2:
        raise DimensionalityError('Projections vector is expected to contain angles')
    # If we've been passed a single vector to map, then add the extra axis
    if len(projections) == 3:
        vector_count = projections.shape[2]
    else:
        projections = projections[..., None]
        vector_count = 1
    
    N = projections.shape[0]
    # from x,y to x,y,z
    out = np.zeros([N, 3, vector_count])
    
    # Ensure unit normals
    vec_mean_normals = normalise_vector(mean_normal)
    
    thetaav = elevation(vec_mean_normals[:, 2])
    phiav = azimuth(vec_mean_normals[:, 0], vec_mean_normals[:, 1])
    phiav, thetaav = clip_angles(phiav, thetaav)
    
    for i in xrange(vector_count):
        # as vector matrix
        kset = projections[:, :, i]
        # find any zero normals as they present a real problem
        # the column indicies are the same in both sets of data
        zero_indices = np.sum(np.abs(kset), axis=1) == 0.0
    
        xs = np.squeeze(kset[:, 0])
        ys = np.squeeze(kset[:, 1])
        # theta, phi
        angles = np.zeros([N, 2])
        
        c = np.sqrt(xs ** 2 + ys ** 2)
        recipc = 1.0 / c
        
        # thetas = asin[cos(c) * sin(thetaav) - (1/c) * yk * sin(c) * cos(thetav)]
        s = np.cos(c) * np.sin(thetaav) + recipc * ys * np.sin(c) * np.cos(thetaav)
        
        # Elevation
        el = np.arcsin(s)
        # Azimuth
        # phis = phiav + atan(psi)
        phis = psi(c, thetaav, xs, ys)
        azi = phiav + np.arctan2(phis[:, 0], phis[:, 1])
        azi, el = clip_angles(azi, el)
        angles[:, 0] = el
        angles[:, 1] = azi
        
        # convert angles to coordinates
        vectors = np.zeros([N, 3])
        vectors[:, 0] = np.cos(angles[:, 1]) * np.sin(angles[:, 0])
        vectors[:, 1] = np.sin(angles[:, 1]) * np.sin(angles[:, 0])
        vectors[:, 2] = np.cos(angles[:, 0])
    
        # reset zero projections back to zero
        vectors[zero_indices, :] = 0.0
        
        # reshape back to column vector
        out[:, :, i] = vectors
        
    return out

def clip_angles(phi, theta):
    # round thetas back in to the range [-pi/2, pi/2]
    pi_over_2 = np.pi / 2.0
    theta[theta > pi_over_2] = theta[theta > pi_over_2] - np.pi
    theta[theta < -pi_over_2] = theta[theta < -pi_over_2] + np.pi
    
    phi[phi > np.pi] = phi[phi > np.pi] - 2.0 * np.pi
    phi[phi < np.pi] = phi[phi < np.pi] + 2.0 * np.pi
    
    return phi, theta

# psi = thetaav != (pi / 2) -> 
#           xk * sin(c) / c * cos(thetaav) * cos(c) - yk * sin(thetav) * sin(c)
#       thetaav == (pi/2)   -> -(xk/yk)
#       thetaav == -(pi/2)  -> xk/yk
def psi(c, thetaav, xs, ys):
    eps = np.spacing(1)
    out = np.zeros([thetaav.size, 2])
        
    neg_pi_ind = np.abs(N.sum(-1) - (np.pi / 2.0)) < eps
    if np.any(neg_pi_ind):
        out[neg_pi_ind, :] = [-xs[neg_pi_ind], -ys[neg_pi_ind]]
    
    pos_pi_ind = np.abs(N.sum(-1) + (np.pi / 2.0)) < eps
    if np.any(pos_pi_ind):
        out[pos_pi_ind, :] = [xs[pos_pi_ind], ys[pos_pi_ind]]
    
    ind = ~np.logical_and(neg_pi_ind, pos_pi_ind)
    out[ind, 0] = xs[ind] * np.sin(c[ind])
    out[ind, 1] = c[ind] * np.cos(thetaav[ind]) * np.cos(c[ind]) - ys[ind] * np.sin(thetaav[ind]) * np.sin(c[ind])
    
    return out
    
# theta = (pi / 2) - asin(nz)
def elevation(zs):
     return (np.pi / 2.0) - np.arcsin(zs) # elevation between 0, pi and 0 at z=1

# phi = atan(ny,nx)
def azimuth(xs, ys):
     return np.arctan2(ys, xs)
    
def normalise_vector(v):
    return v / np.sqrt((v ** 2).sum(-1))[..., None]
```

```python
N = normalise_vector(np.random.random([10000, 3]))
bp = normalise_vector(np.random.random([10000, 3]))
```

```python
snape = expmap_snape(logmap_snape(N, bp), bp)
smith = expmap(logmap(N, bp), bp)
print np.allclose(np.squeeze(snape), N)
print np.allclose(np.squeeze(smith), N)
np.allclose(np.squeeze(snape), np.squeeze(smith))
```
